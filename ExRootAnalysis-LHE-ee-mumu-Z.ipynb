```python
dir__ = 'SM_ee_mumu_80GeV'
path__ = 'data/{}/run_01/unweighted_events.root'.format( dir__ )
df = read_lhe_uproot( path__ )
df.head(50)
```

```python
def process_select_data( df ):
    df__ = df
    df__[ 'cos_theta' ] = df__.Pz / np.sqrt( df__.Pz ** 2 + df__.PT ** 2 )
    selection__ = ( ( df__.Status == 1 ) & ( df__.PID == 13 ) )
    # selection = ( ( df.Status == 1 ) & ( df.PID == -13 ) )
    # selection = ( ( df.Status == 1 ) & ( df.PID == 11 ) )
    df_sel__ = df__[ selection__ ]
    return df_sel__
```

```python
df_sel = process_select_data( df )
df_sel
```

```python
# 80 GeV
cross_section = 3.512700e+01
```

```python
n_events = len( df_sel )
print ( n_events )
weights = np.ones( n_events )
weights
```

```python
def plot_cos_theta( dfsel, weights, label='', ylabel='', bins=50, range=(-1.,1.), linestyle='solid' ):
    dfsel__ = dfsel
    label__ = label
    bins_ = bins
    range_ = range
    linestyle__ = linestyle
#     plt.hist( dfsel__[ 'cos_theta' ], histtype='step', bins=100, label=label__, linestyle=linestyle__ )
    plt.hist( dfsel__[ 'cos_theta' ], weights=weights, histtype='step', bins=bins_, range=range_, label=label__, linestyle=linestyle__ )
    plt.xlabel( r"$\cos\theta$" )
    plt.ylabel( ylabel )
```

```python
fig = plt.figure( figsize=(12,10) )
plot_cos_theta( df_sel, weights=weights, bins=50, ylabel="Número de eventos" )
```

```python
print ( cross_section )
weights_sigma = np.ones( n_events ) * ( cross_section / n_events )
weights_sigma
```

```python
fig = plt.figure( figsize=(12,10) )
plot_cos_theta( df_sel, weights=weights_sigma, bins=50, ylabel=r"$\Delta \sigma_i\,(\rm{pb})$" )
```

### Seção de choque *diferencial*

```python
range_min = -1.
range_max = +1.
bins = 50
delta_cos_theta = ( ( range_max - range_min ) / bins )
print ( delta_cos_theta )

weights_binwidth = weights_sigma / delta_cos_theta
print ( weights_binwidth )

fig = plt.figure( figsize=(12,10) )
plot_cos_theta( df_sel,
                weights=weights_binwidth,
                bins=bins, range=( range_min, range_max ),
                ylabel=r"$\frac{\Delta \sigma_i}{\Delta(\cos\theta)} \approx \frac{d\sigma}{d\cos\theta} \,(\rm{pb})$" )
```

### Número de eventos esperados (para um valor de luminosidade/fluxo)

```python
L = 100. # pb^{-1}
print ( "cross section = {} (pb)".format( cross_section ) )
print ( "L = {} (1/pb)".format( L ) )

N_esp = cross_section * L
print ( "N(esp.) = {:.2f}".format( N_esp ) )
```

```python
print ( cross_section )
weights_nexpected = np.ones( n_events ) * ( cross_section * L / n_events )
print ( weights_nexpected )

fig = plt.figure( figsize=(12,10) )
plot_cos_theta( df_sel, weights=weights_nexpected, bins=50, ylabel="Número esperado de eventos (L = {:.0f}{})".format( L , r"$\rm{pb}^{-1}$" ) )
```
