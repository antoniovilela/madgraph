```python
df = {}
for label__ in [ '20GeV', '40GeV', '60GeV', '80GeV', 'MZ', '100GeV' ]:
    dir__ = 'SM_ee_mumu_{}'.format( label__ )
    path__ = 'data/{}/run_01/unweighted_events.root'.format( dir__ )
    df[ label__ ] = read_lhe_uproot( path__ )
    
df['MZ']
```

```python
def process_select_data( df ):
    df__ = df
    df__[ 'cos_theta' ] = df__.Pz / np.sqrt( df__.Pz ** 2 + df__.PT ** 2 )
    selection__ = ( ( df__.Status == 1 ) & ( df__.PID == 13 ) )
    # selection = ( ( df.Status == 1 ) & ( df.PID == -13 ) )
    # selection = ( ( df.Status == 1 ) & ( df.PID == 11 ) )
    df_sel__ = df__[ selection__ ]
    return df_sel__
```

```python
df_sel = {}
for label__ in [ '20GeV', '40GeV', '60GeV', '80GeV', 'MZ', '100GeV' ]:
    df_sel[ label__ ] = process_select_data( df[ label__ ] )
    
df_sel['MZ']
```

```python
cross_section = {}
cross_section[ '20GeV' ]  = ( 2.244000e+02, 9.232139e-02 )
cross_section[ '40GeV' ]  = ( 5.646700e+01, 2.709858e-02 )
cross_section[ '60GeV' ]  = ( 2.680500e+01, 1.535268e-02 )
cross_section[ '80GeV' ]  = ( 3.512700e+01, 1.467866e-02 )
cross_section[ 'MZ' ]     = ( 2.117100e+03, 7.280896e-01 )
cross_section[ '100GeV' ] = ( 5.313900e+01, 1.782526e-02 )
```

```python
fig = plt.figure( figsize=(12,10) )
X__ = ( 20., 40., 60., 80., 91.188, 100. )
Y__ = ( cross_section[ '20GeV' ][ 0 ],
        cross_section[ '40GeV' ][ 0 ],
        cross_section[ '60GeV' ][ 0 ],
        cross_section[ '80GeV' ][ 0 ],
        cross_section[ 'MZ' ][ 0 ],
        cross_section[ '100GeV' ][ 0 ]
      )
plt.plot( X__, Y__, 'ko' )
plt.xlabel( r"$\sqrt{s}\,(\rm{GeV})$" )
plt.ylabel( r"$\sigma\,(\rm{pb})$" )
plt.yscale( 'log' )
```

```python
def plot_cos_theta( dfsel, label='', linestyle='solid' ):
    dfsel__ = dfsel
    label__ = label
    linestyle__ = linestyle
    plt.hist( dfsel__[ 'cos_theta' ], histtype='step', bins=100, label=label__, linestyle=linestyle__ )
    plt.xlabel( r"$\cos\theta$" )
    plt.ylabel( "Número de eventos" )
```

```python
fig = plt.figure( figsize=(12,10) )
label__ = '20GeV'
plot_cos_theta( df_sel[ label__ ], label=label__ )
label__ = '40GeV'
plot_cos_theta( df_sel[ label__ ], label=label__ )
label__ = '60GeV'
plot_cos_theta( df_sel[ label__ ], label=label__ )
label__ = '80GeV'
plot_cos_theta( df_sel[ label__ ], label=label__ )
label__ = 'MZ'
plot_cos_theta( df_sel[ label__ ], label=label__ )
label__ = '100GeV'
plot_cos_theta( df_sel[ label__ ], label=label__ )

plt.legend( loc='best' )
```

```python
fig = plt.figure( figsize=(12,10) )
for label__ in [ '20GeV', '40GeV', '60GeV', '80GeV', 'MZ', '100GeV' ]:
    linestyle__ = 'solid'
    if label__ == 'MZ': linestyle__ = 'dashed'
    plot_cos_theta( df_sel[ label__ ], label=label__, linestyle=linestyle__ )
plt.legend( loc='best' )
```
